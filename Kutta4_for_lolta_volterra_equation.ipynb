```
import matplotlib.pyplot as plt

import numpy as np
import cmath as cm
```

```
def RK4(fcn, Tspan, I, Nt, r, d, y, gam):
  t = np.linspace(Tspan[0], Tspan[1], Nt+1)
  dt = t[1] - t[0]
  f_ = lambda u, t, r, d, y, gam: np.asarray(fcn(u, t, r, d, y, gam))
  if isinstance(I, (float, int)):
    u = np.zeros(Nt+1) # u[k] is the numerical solution
  else:
    I = np.asarray(I)
    neq = I.size
    u = np.zeros((Nt+1, neq))
  u[0] = I
  for n in range(Nt):
    k1 = dt*f_( u[n], t[n], r, d, y, gam)
    k2 = dt*f_( u[n] + 0.5*k1, t[n] + 0.5*dt,r, d, y, gam)
    k3 = dt*f_( u[n] + 0.5*k2, t[n] + 0.5*dt, r, d, y, gam)
    k4 = dt*f_( u[n] + k3, t[n+1],r, d, y, gam)
    u[n+1] = u[n] + ( k1 + 2.0 * ( k2 + k3 ) + k4 ) / 6.0
  return u, t
```

```
def fcn(X,t, r, d, y, gam):
  Nprime = r*X[0] -   y*X[0]*X[1]
  Pprime = -d*X[1] + gam*y*X[0]*X[1]
  return [Nprime,Pprime]
```

```
Tspan = [0,200]
Nt = 1000

Nprime0= 90; Pprime0=100
I = [Nprime0,Pprime0]
# Definition of parameters
r = 0.2 # random values of parameters
y = 0.005
d = 0.2
gam = 1
u, t = RK4(fcn, Tspan, I, Nt, r, d, y, gam)
#plt.plot(t, u[:,0])#,linestyle=lstyle[counter],label="a="+str(a),linewidth=1.5)
#plt.title(r'$K =$ '+str(k+2)+', '+r'$\omega = 1.5$ ',fontsize=25)
#counter=counter+1
#plt.ylim((0,12))
#plt.xlabel(r'$t$',fontsize=22)
#plt.ylabel(r'$M$',fontsize=22)
#plt.legend(loc='best', fontsize=14)
#plt.savefig('M-t2.eps',dpi = 10000)
#plt.show()
plt.figure(figsize =(10,10) )
plt.plot(t,u[:,1],linewidth=1.5,label=r'P(t)')
plt.plot(t,u[:,0],linewidth=1.5,label=r'n(t)')
#plt.plot(u[:,1],u[:,0],linewidth=1.5,label=r'n(t)')
#plt.ylim((0,12))
plt.xlabel(r'$t$',fontsize=22)
plt.legend(loc='best', fontsize=14)
plt.grid()
plt.show()
```
